```python
import numpy as np
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc("font", size=12)
plt.rcParams["figure.figsize"] = (20, 8)
```

```python
df = pd.read_csv("time_results.csv")
df.head()
```

```python
import re

method_new_line = [re.sub('(::)|[ _]' , "\n", method) for method in df["method"].values]
df = df.assign(methodNL=method_new_line)
df
```

```python
width = 0.5

plt.bar(df[df["lang"] == "php"]["methodNL"], df[df["lang"] == "php"]["elapsed-time(sec)"], width,label="php", color="purple")
plt.bar(df[df["lang"] == "java"]["methodNL"], df[df["lang"] == "java"]["elapsed-time(sec)"], width, label="java", color="red")
plt.bar(df[df["lang"] == "cpp"]["methodNL"][:3], df[df["lang"] == "cpp"]["elapsed-time(sec)"][:3], width, label="cpp [NO OPT]", color="green")
plt.bar(df[df["lang"] == "cpp"]["methodNL"][3:], df[df["lang"] == "cpp"]["elapsed-time(sec)"][3:], width, label="cpp", color="blue")
plt.title("Word count times for different prog langs, using different approaches to do it")
plt.ylabel("Elapsed time (sec)")
plt.xlabel("The approach or data structure used to count words")
plt.legend()
```
